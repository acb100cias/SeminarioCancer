```python
def dis_lap(M):
  L=-4*M
  L += np.roll(M,(0,-1),(0,1))
  L += np.roll(M,(0,1),(0,1))
  L += np.roll(M,(-1,0),(0,1))
  L += np.roll(M,(1,0),(0,1))
  return L
```

```python
dis_lap(A)
```

```python
def gray_scott(A,B,DA,DB,f,k,delta_t):
  LA=dis_lap(A)
  LB=dis_lap(B)
  dvA=(DA*LA-A*B**2+f*(1-A))*delta_t
  dvB=(DB*LB+A*B**2-(k-f)*B)*delta_t
  A+=dvA
  B+=dvB
  return A,B
```

```python
def ci(N,ruido=0.2):
  A=(1-ruido)*np.ones((N,N))+ ruido*np.random.random((N,N))
  B=ruido*np.random.random((N,N))
  N2= N//2
  r=int(N/10.)
  A[N2-r:N2+r,N2-r:N2+r]=0.5
  B[N2-r:N2+r,N2-r:N2+r]=0.25
  return A,B
```

```python
def draw(A,B):
  fig,ax=plt.subplots(1,2,figsize=(6,4))
  ax[0].imshow(A,cmap='gray')
  ax[1].imshow(B,cmap='gray')
  ax[0].set_title('A')
  ax[1].set_title('B')
  ax[0].axis('off')
  ax[1].axis('off')

```

```python
A,B=ci(200)
draw(A,B)
```

```python
delta=1.
DA=0.16
DB=0.08
f=0.06
k=0.062
N=400
pasos=600
```

```python
A,B=ci(N)
for t in range(pasos):
  A,B=gray_scott(A,B,DA,DB,f,k,delta)
  draw(A,B)
```
